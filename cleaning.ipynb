```python
import pandas as pd
import numpy as np
import re
from indo_normalizer import Normalizer
```

```python
df = pd.read_csv(r"data_klien_1\Doctor QnA Indonesia.csv")
df.info()
```

```python
df.head()
```

```python
df_cleaned = df.copy()

def clean_text(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Remove URLs
    text = re.sub(r'\@\w+|\#','', text)  # Remove mentions and hashtags
    text = re.sub(r'[^A-Za-z0-9\s.,]+', '', text)  # Remove special characters
    text = text.lower()  # Convert to lowercase
    text = text.strip()  # Remove leading/trailing whitespace
    return text
df_cleaned['question'] = df_cleaned['question'].apply(clean_text)
df_cleaned['answer'] = df_cleaned['answer'].apply(clean_text)
```

```python
save_path = r"data_klien_1\Doctor_QnA_Indonesia_Cleaned.csv"
df_cleaned.to_csv(save_path, index=False)
```
